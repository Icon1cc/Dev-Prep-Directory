## Topic: Linear Regression with Normal Equation
Solve for linear model weights in closed form.

```python
import numpy as np

# X: input (n_samples, n_features)
X = np.array([[1, 1], [1, 2], [1, 3], [1, 4]], dtype=float)

# y: target (n_samples,)
y = np.array([2, 3, 4, 5], dtype=float)

w = np.linalg.inv(X.T @ X) @ X.T @ y
predictions = X @ w

print("weights:", w)
print("predictions:", predictions)
```

## Topic: Principal Component Analysis
Reduce feature dimensions using manual PCA steps and sklearn PCA.

```python
from sklearn.decomposition import PCA

#Manually
X_centered = X - X.mean(axis=0)
cov_matrix = (X_centered.T @ X_centered) / len(X)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Top 2 components
X_pca = X_centered @ eigenvectors[:, :2]

# Or using sklearn

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print("PCA result:\n", X_pca)
```

## Topic: Cosine Similarity
Compare manual and sklearn cosine similarity implementations.

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def cosine_similarity_manual(vec1, vec2):
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

vec_a = np.array([1, 2, 3], dtype=float)
vec_b = np.array([4, 5, 6], dtype=float)

sim_manual = cosine_similarity_manual(vec_a, vec_b)
sim_sklearn = cosine_similarity([vec_a], [vec_b])[0][0]

print(f"Manual cosine similarity: {sim_manual}")
print(f"Sklearn cosine similarity: {sim_sklearn}")
```

## Topic: Gradient Descent for Linear Regression
Optimize linear model weights iteratively using gradients.

```python
import numpy as np

# Simple gradient descent
X_gd = np.array([[1, 1], [1, 2], [1, 3], [1, 4]], dtype=float)
y_gd = np.array([2, 3, 4, 5], dtype=float)

w_gd = np.zeros(X_gd.shape[1], dtype=float)
learning_rate = 0.01

for epoch in range(1000):
    predictions = X_gd @ w_gd
    error = predictions - y_gd
    gradient = X_gd.T @ error / len(X_gd)
    w_gd -= learning_rate * gradient

print("Learned weights:", w_gd)
print("Final predictions:", X_gd @ w_gd)
```

## Topic: Axis-Wise Aggregations
Compute sums, means, max, and min along rows or columns with axis.

```python
X = np.array([[1, 2, 3],
              [4, 5, 6]])

# Sum along axis 0 (down rows)
col_sums = X.sum(axis=0)  # [5, 7, 9]

# Sum along axis 1 (across columns)
row_sums = X.sum(axis=1)  # [6, 15]

# Mean, max, min work similarly
col_means = X.mean(axis=0)
col_maxes = X.max(axis=0)
col_mins = X.min(axis=0)

print("Column sums:", col_sums)
print("Row sums:", row_sums)
print("Column means:", col_means)
print("Column maxes:", col_maxes)
print("Column mins:", col_mins)
```
